```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sodapy import Socrata
import os.path
```

```python
def pull_data():
    
    if os.path.isfile("data/case_data.pkl"):
        results_df = pd.read_pickle("data/case_data.pkl")
    else:
        # API usage: https://dev.socrata.com/foundry/data.cdc.gov/vbim-akqf
        client = Socrata("data.cdc.gov", None)
        results = client.get("vbim-akqf", where="cdc_case_earliest_dt > '2021-07-01T00:00:00.000'", limit=10000000)
        results_df = pd.DataFrame.from_records(results)

        # save a copy
        results_df.to_pickle("data/case_data.pkl")
        
    return results_df
```

```python
results_df = pull_data()
```
